```python
def intro():
    Name = input('Please provide your name')
    print(f'Welcome to Madam\'B Kichen {Name}')
    print('Home of delicious meal')
    print('We are always at your service')
    print('Please sit on any row to be served')
def main():
    intro()
    tag = int(input('choice a tag and proceed to serve you meal:'))
    if tag == 1:
        plate = int(input('Numbers of plate:'))
        rice = int(input('Portion of rice:'))
        chicken = int(input('Portion of chicken:'))
        salad = int(input('Portion of salad:'))
        plantain = int(input('Portion of plantain:'))
        water = int(input('Bottle of water:'))
        drink = int(input('Portion of Drink:'))
        
        total_rice = rice * 500
        total_chicken = chicken * 1200
        total_salad = salad * 700
        total_plantain = plantain * 300
        total_water = water * 200
        total_drink = drink * 300
        
        total_price = plate*(total_rice + total_chicken + total_salad + total_plantain + total_water + total_drink)
        print(f'Find above your order and the total price is {total_price}')
        print('Thanks for visiting Resturant')
        feedback = input('Give us a rating and feedback')
    elif tag == 2:
        plate = int(input('Numbers of plate:'))
        swallow_with_any_soup_of_your_choice = int(input('Raps of swallow:'))
        chicken = int(input('Portion of chicken:'))
        cow_head = int(input('Portion of cow head:'))
        beef = int(input('Portion of beef:'))
        goat_head = int(input('Portion of goat head:'))
        water = int(input('Bottle of water:'))
        drink = int(input('Portion of Drink:'))
        
        total_swallow= swallow_with_any_soup_of_your_choice * 500
        total_chicken = chicken * 1200
        total_cow_head = cow_head * 100
        total_beef = beef * 800
        total_goat_head = goat_head * 700
        total_water = water * 200
        total_drink = drink * 300
        
        total_price = plate*(total_swallow + total_chicken + total_cow_head + total_beef + total_goat_head + total_water + total_drink)
        print(f'Find above your order and the total price is {total_price}')
        print('Thanks for visiting Resturant')
        feedback = input('Give us a rating and feedback')
    elif tag == 3:
        plate = int(input('Numbers of plate:'))
        meatpie = int(input('portion of meatpie:'))
        chicken_and_chips = int(input('Portion of chicken_and_chips:'))
        shawama = int(input('Portion of shawama:'))
        burger = int(input('Portion of burger:'))
        cake = int(input('Portion of cake:'))
        water = int(input('bottle of water:'))
        drink = int(input('Portion of Drink:'))
        
        total_meatpie= meatpie * 450
        total_chicken_chips = chicken_and_chips * 2500
        total_shawama = shawama * 1500
        total_burger = burger * 700
        total_cake = cake * 1500
        total_water = water * 200
        total_drink = drink * 300
        
        total_price = plate*(total_meatpie + total_chicken_chips + total_shawama  + total_burger + total_cake + total_water + total_drink)
        print(f'Find above your order and the total price is {total_price}')
        print('Thanks for visiting Resturant')
        feedback = input('Give us a rating and feedback:')
    else:
        print('Invalid Tag')
main()
```
